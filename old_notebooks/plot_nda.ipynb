```python
import sys
sys.path.insert(1, '../') # make sure to use the code in this repo
import os
os.environ['CDF_LIB'] = '/home/peijin/cdf/cdf38_0-dist/lib'
import datetime
import numpy as np
import pandas as pd
from spacepy import pycdf
from astropy.io import fits as pyfits
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter
# myFmt = DateFormatter('%Y/%m/%d %H:%M')
myFmt = DateFormatter('%H:%M')
import matplotlib.dates as mdates
import matplotlib as mpl
# try to use the precise epoch 
mpl.rcParams['date.epoch'] = '1970-01-01T00:00:00'
try:
    mdates.set_epoch('1970-01-01T00:00:00')
except:
    pass
import warnings
warnings.filterwarnings('ignore')
```

```python
### define x and y axes 
nda_freq = pd.DataFrame(nda_data[3].data).values.reshape(-1)
nda_time = pd.DataFrame(nda_data[4].data).values.reshape(-1)
nda_arr1 = nda_data[1].data
nda_arr2 = nda_data[2].data

datetime_obj_start = nda_data[0].header['DATE-OBS'] + ' ' + nda_data[0].header['TIME-OBS']
datetime_obj_end = nda_data[0].header['DATE-OBS'] + ' ' + nda_data[0].header['TIME-END']

start_obs_time = datetime.datetime.strptime(datetime_obj_start, '%d/%m/%Y %H:%M:%S')
end_obs_time = datetime.datetime.strptime(datetime_obj_end, '%d/%m/%Y %H:%M:%S')
```

```python
plt.figure(figsize=(8,5), dpi=100)
plt.imshow(nda_arr1, 
           aspect='auto', origin='lower', 
           #vmin=(np.mean(nda_arr)-2 * np.std(nda_arr)), 
           #vmax=(np.mean(nda_arr)+3 * np.std(nda_arr)), 
           extent=[mdates.date2num(start_obs_time), mdates.date2num(end_obs_time), nda_freq[0], nda_freq[-1]], 
           cmap='inferno')

plt.xlabel('Time (UT)')
plt.ylabel('Frequency (MHz)')
plt.title(nda_data[0].header['ORIGIN'] + ' [1] ' + nda_data[0].header['DATE-OBS'])
plt.gca().xaxis.set_major_formatter(myFmt)
plt.tight_layout()
```

```python
plt.figure(figsize=(8,5), dpi=100)
plt.imshow(nda_arr2, 
           aspect='auto', origin='lower', 
           #vmin=(np.mean(nda_arr)-2 * np.std(nda_arr)), 
           #vmax=(np.mean(nda_arr)+3 * np.std(nda_arr)), 
           extent=[mdates.date2num(start_obs_time), mdates.date2num(end_obs_time), nda_freq[0], nda_freq[-1]], 
           cmap='inferno')

plt.xlabel('Time (UT)')
plt.ylabel('Frequency (MHz)')
plt.title(nda_data[0].header['ORIGIN'] + ' [2] ' + nda_data[0].header['DATE-OBS'])
plt.gca().xaxis.set_major_formatter(myFmt)
plt.tight_layout()
```

```python
plt.figure(figsize=(8,5), dpi=100)
plt.imshow(nda_arr2 - nda_arr1, 
           aspect='auto', origin='lower', 
           # vmin=np.percentile(nda_arr2 - nda_arr1, 0.90), 
           # vmax=np.percentile(nda_arr2 - nda_arr1, 0.95), 
           extent=[mdates.date2num(start_obs_time), mdates.date2num(end_obs_time), nda_freq[0], nda_freq[-1]], 
           cmap='inferno')

plt.colorbar()
plt.xlabel('Time (UT)')
plt.ylabel('Frequency (MHz)')
plt.title('Difference: ' + nda_data[0].header['DATE-OBS'])
plt.gca().xaxis.set_major_formatter(myFmt)
plt.tight_layout()
```
